## Explore relationships between numerical variables (optional)

### Subtask:
Explore relationships between numerical variables within and across the two datasets. If meaningful relationships are expected between numerical variables within or across the two datasets, consider creating scatter plots or correlation matrices to visualize these relationships.


**Reasoning**:
Identify numerical columns for correlation analysis, handle placeholder values by replacing them with NaN, merge the two dataframes on 'datetime' and station ID, calculate the correlation matrix, and visualize it using a heatmap.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical columns for correlation analysis (excluding identifiers that aren't meaningful for correlation)
numerical_cols_weather_corr = df_weather.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_cols_pm10_corr = df_pm10.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Exclude 'TM' and station ID columns from direct correlation calculation if not needed
if 'TM' in numerical_cols_weather_corr:
    numerical_cols_weather_corr.remove('TM')
# Removed 'STN' from exclusion as it's needed for merging
if 'TM' in numerical_cols_pm10_corr:
    numerical_cols_pm10_corr.remove('TM')
# Removed 'STN_ID' from exclusion as it's needed for merging

# Handle placeholder values (-9, -99) by replacing them with NaN
df_weather_cleaned = df_weather.replace([-9, -99], np.nan)
df_pm10_cleaned = df_pm10.replace([-9, -99], np.nan)

# Convert 'datetime' columns to datetime objects if they are not already
df_weather_cleaned['datetime'] = pd.to_datetime(df_weather_cleaned['Date'] + ' ' + df_weather_cleaned['Time'])
df_pm10_cleaned['datetime'] = pd.to_datetime(df_pm10_cleaned['Date'] + ' ' + df_pm10_cleaned['Time'])

# Ensure the station ID column names are consistent for merging
# The STN column in df_weather needs to be renamed to STN_ID to match df_pm10
df_weather_cleaned.rename(columns={'STN': 'STN_ID'}, inplace=True)

# **Debugging Merge Issue:** Check unique values and data types of merge keys
print("df_weather_cleaned 'STN_ID' unique values:", df_weather_cleaned['STN_ID'].unique())
print("df_pm10_cleaned 'STN_ID' unique values:", df_pm10_cleaned['STN_ID'].unique())
print("df_weather_cleaned 'datetime' data type:", df_weather_cleaned['datetime'].dtype)
print("df_pm10_cleaned 'datetime' data type:", df_pm10_cleaned['datetime'].dtype)
print("df_weather_cleaned 'datetime' first 5 values:", df_weather_cleaned['datetime'].head())
print("df_pm10_cleaned 'datetime' first 5 values:", df_pm10_cleaned['datetime'].head())


# Merge the dataframes on 'datetime' and station ID
merged_df = pd.merge(df_weather_cleaned, df_pm10_cleaned[['STN_ID', 'datetime', 'PM10']], on=['STN_ID', 'datetime'], how='inner')

# Select numerical columns from the merged dataframe for correlation analysis
# Include 'PM10' and relevant weather parameters
selected_numerical_cols = [col for col in numerical_cols_weather_corr if col in merged_df.columns] + ['PM10']
merged_numerical_df = merged_df[selected_numerical_cols]

# Check if the merged dataframe is empty
if merged_numerical_df.empty:
    print("Merged dataframe is empty. Cannot perform correlation analysis or generate scatter plots.")
else:
    # Calculate the correlation matrix
    correlation_matrix = merged_numerical_df.corr(numeric_only=True)

    # Visualize the correlation matrix using a heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix of Numerical Variables')
    plt.show()

    # Select pairs of numerical variables that show potentially interesting relationships
    # Based on the heatmap, let's select pairs with absolute correlation above a certain threshold (e.g., 0.3)
    # Use .stack() and .unstack() to handle potential MultiIndex from corr()
    stacked_corr = correlation_matrix.stack()
    interesting_correlations = stacked_corr.sort_values(kind="quicksort")

    # Filter out self-correlations and duplicate pairs
    interesting_correlations = interesting_correlations[interesting_correlations != 1.0]
    interesting_correlations = interesting_correlations[abs(interesting_correlations) >= 0.3]
    # Get every second value to avoid duplicate pairs - this assumes a symmetric matrix structure after filtering
    interesting_correlations = interesting_correlations.iloc[::2]


    print("Potentially interesting correlations (absolute value >= 0.3):")
    print(interesting_correlations)

    # Create scatter plots for the selected pairs of variables
    for (var1, var2), corr_value in interesting_correlations.items():
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=merged_numerical_df[var1], y=merged_numerical_df[var2])
        plt.title(f'Scatter Plot of {var1} vs {var2} (Correlation: {corr_value:.2f})')
        plt.xlabel(var1)
        plt.ylabel(var2)
        plt.show()
```
